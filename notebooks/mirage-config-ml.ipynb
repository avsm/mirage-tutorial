```ocaml
open Lwt
module Unikernel = struct
    module Main (C: V1_LWT.CONSOLE) = struct
    
        let start c =
            for_lwt i = 0 to 4 do
                C.log c "hello" ;
                lwt () = OS.Time.sleep 1.0 in
                C.log c "world" ;
                return ()
        done
    end
end
```

```ocaml
open Mirage

let main = foreign "Unikernel.Main" (console @-> job)

let jobs = [
    main $ default_console
    ]
```

```ocaml
open Lwt

let _ = Printexc.record_backtrace true

let console1 () =
  Console.connect "0"

module M1 = Unikernel.Main(Console)

let t1 () =
  console1 () >>= function
  | `Error e -> fail (Failure "console1")
  | `Ok console1 ->
  M1.start console1
```

```ocaml
let () =
  OS.Main.run (join [t1 ()])
```
