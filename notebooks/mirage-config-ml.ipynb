# Mirage `hello world`: Building a Unikernel 

Now you understand the basics of `Lwt`, the threading library used throughout mirage, let's go ahead and build a simple `hello world` unikernel. This will use only the `console` device driver to fire off some threads that output some text.

Before we begin, let's load the necessary Mirage libraries. First, the Mirage library itself plus the necessary module type definitions that allow us to swap backends by swapping compatible implementations of libraries:

```ocaml
#require "mirage-types,mirage,lwt.syntax,mirage-types.lwt"
```

Then, for the purposes of this `iocaml` notebook, let's load the necessary Mirage implementation libraries to give us basic Unix support and console access:

```ocaml
#require "mirage-console-unix"
```

Building a Mirage application then consists of the following steps:

1. Define the Unikernel behaviour, adhering to the necessary module type signatures; this means that devices are made available to the unikernel as functor parameters.

2. Write a `config.ml` file that describes the structure of the unikernel in terms of the devices that will be made available.

3. Configure the unikernel by invoking the Mirage command line tool with appropriate parameters; this generates the necessary `Makefile` plus the `main.ml` that 

4. Build the unikernel using the autogenerated `Makefile`.

We will now go through these stages in turn.

## 1. Defining a Unikernel

First, the code:

```ocaml
module Unikernel = struct
    open Lwt

    module Main (C: V1_LWT.CONSOLE) = struct
    
        let start c =
            for_lwt i = 0 to 4 do
                C.log c "hello";
                OS.Time.sleep 1.0 >>= fun () ->
                C.log c "world";
                return ()
        done
        
    end
end
```

To veteran OCaml programmers among you, this might look a little odd: we have a `Main` module parameterised by another module (`C`, of type `CONSOLE`) that contains a method `start` taking a single parameter `c` (an instance of a `CONSOLE`). This is the basic structure required to make this a Mirage unikernel rather than a standard OCaml POSIX application.

The concrete implementation of `CONSOLE` will be supplied at compile-time, depending on the target that you are compiling for.  This configuration is stored in `config.ml`, which is very simple for our first application.

## 2. Write a `config.ml`

The `config.ml` specifies a set of **jobs**, equivalent to top-level threads with attached devices, that represent processes with start/stop lifecycle. In this case it is rather simple as the only device depended upon is the `console`:

```ocaml
module Config = struct
    open Mirage
    
    let main = 
        foreign "Unikernel.Main" (console @-> job)
    
    let jobs = [
        main $ default_console
        ]
    
    (* This completes the `config.ml` by giving an entry point where
       `jobs` are registered, effectively to specify the the top-level 
       `Lwt` threads that are invoked at start-of-day. Limitations of 
       current support in Mirage for IOCaml requires this be commented
       out for now. 
    
    let () = 
        register "console" jobs
    *)
end
```

The configuration file is a normal OCaml module that calls `register` to create one or more jobs, each of which represent a process (with a start/stop lifecycle).  Each job most likely depends on some device drivers; all the available device drivers are [defined in the `Mirage` module](http://mirage.github.io/mirage).

In this case, the `main` variable declares that the entry point of the process is the `Main` module from the file `unikernel.ml`.  The `@->` combinator is used to add a device driver to the list of functor arguments in the job definition (see `unikernel.ml`), and the final value of using this combinator should always be a `job` if you intend to register it.

Notice that we refer to the module name as a string here, instead of directly as an OCaml value.  The `mirage` command-line tool evaluates this configuration file at build-time and outputs a `main.ml` that has the concrete values filled in for you, with the exact modules varying by which backend you selected (e.g. Unix or Xen).

Mirage mirrors the Xen model on UNIX as far as possible: your application is built as a unikernel which needs to be instantiated and run whether on UNIX or on Xen. When your unikernel is run, it starts much as a VM on Xen does -- and so must be passed references to devices such as the console, network interfaces and block devices on startup.

In this case, this simple `hello world` example requires just a console for output, so we register a single `Job` consisting of the `Unikernel.Main` module (and, implicitly its `start` function) and passing it a single reference to a console.

You can find the module signatures of all the device drivers (such as `CONSOLE`) in the [`types/`](https://github.com/mirage/mirage/tree/master/types) directory of the main Mirage repository.  Since you'll find yourself referring back to these quite often when building Mirage applications, it's worth bookmarking the [documentation](http://mirage.github.io) for this module.

## 3. Generate `main.ml` and `Makefile`

Invoking the Mirage command line tool with appropriate parameters uses the `config.ml` to generate the necessary code in `main.ml` to define the actual entry point of your unikernel:

```ocaml
Sys.command 
    "mirage configure ../mirage-skeleton/console/config.ml --unix"
```

This results in:

+ a `Makefile` to manage building the unikernel:
    + `make depend` to install dependencies via the `opam` package manager;
    + `make build` to compile the unikernel; and
    + `make run` to run the unikernel if it was configured for Unix.
    
  Running on Xen requires invoking your Xen toolchain of choice (e.g., `xm` or `xl`) to boot the resulting VM (`.xen`) according to the generated `.xl` Xen configuration file.
  
+ a `main.ml` that actually provides the necessary entrypoint to your unikernel.

If we examine the `main.ml` we see that it involves, first, constructing and then hotplugging the necessary devices (in this case, only the console):

```ocaml
(* Generated by Mirage (Tue, 29 Jul 2014 22:24:55 GMT). *)

open Lwt
    
module Main = struct
    let _ = Printexc.record_backtrace true
    
    let console1 () =
      Console.connect "0"
    
    module M1 = Unikernel.Main(Console)
    
    let t1 () =
      console1 () >>= function
      | `Error e -> fail (Failure "console1")
      | `Ok console1 ->
      M1.start console1
end
```

Followed by, finally, actually invoking the constructed "processes" (that is, `Lwt` threads) -- in this case, there is only one, `t1`:

```ocaml
let () =
  OS.Main.run (join [Main.t1 ()])
```

## 4. Build the Unikernel

In this IOCaml notebook, it is enough to invoke the code above; in practice you would invoke the generated `Makefile` appropriately to build and potentially run your unikernel:

```ocaml
Sys.command 
    "make depend && make build"
```

Rebuilding your unikernel for Xen is then simply a matter of reconfiguring appropriately, and repeating the steps above. Note that building for Xen requires you to be running Mirage on a Linux platform (i.e., not OSX):

```ocaml
Sys.command
    "mirage configure ../mirage-skeleton/console/config.ml --xen"
```

```ocaml
Sys.command
    "make depend && make build"
```
