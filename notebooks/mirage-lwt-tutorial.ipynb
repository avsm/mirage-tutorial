# Getting started with `Lwt` threads

Lwt is a lightweight cooperative threading library for OCaml. A good way to understand Lwt and its use in Mirage is to write some simple code. This notebook introduces the basic concepts alongside some simple exercises. Example code for all exercises is in the [mirage-skeleton/lwt/src](https://github.com/mirage/mirage-skeleton/tree/master/lwt/src) repository.

Note that, although Lwt has several syntax extensions that have been widely used in Mirage in the past, current idiomatic usage is to avoid the syntax extensions in favour of direct operators. 

## Concepts

Here we give a minimal introduction to Lwt. See the [Lwt manual](http://ocsigen.org/lwt/manual/) for full details. 

### `return`

The first useful function is the `return` statement, which constructs a constant thread:

```ocaml
  val return: 'a -> 'a Lwt.t
```

It is used to construct a thread that immediately returns with the provided value. Once the value is wrapped in an Lwt thread, it cannot directly be used (as the thread may not have completed yet). 

### `bind`

This is where `bind` comes in:

```ocaml
  val bind: 'a Lwt.t -> ('a -> 'b Lwt.t) -> 'b Lwt.t
```

`bind t f` creates a thread which waits for `t` to terminate, then passes the result to `f`. If `t` is a sleeping thread, then `bind t f` will be a sleeping thread too, until `t` terminates. If `t` fails, then the resulting thread will fail with the same exception.

### `join` / `choose`

There are two important operations to compose threads: `join` and `choose`.

```ocaml
  val join : unit Lwt.t list -> unit Lwt.t
```

`join` takes a list of threads and waits for all of them to terminate. If at least one thread fails then `join l` will fail with the same exception as the first to fail, after all threads terminate.

```ocaml
 val choose : 'a t list -> 'a t
```

`choose l` behaves as the first thread in l to terminate. If several threads are already terminated, one is chosen at random.

The infix operators `<&>` and `<?>` are defined in the Lwt module, where `a <&> b` is equivalent to `join [a; b]`, and `a <?> b` is equivalent to `choose [a; b]`.

### Core Event Loop

The core of Lwt is based on an event loop. In "standard" (non-Mirage) settings, this loop is started using the `Lwt_main.run` function. However, when using Mirage, the loop is automatically started by the `main.ml` file autogenerated by the mirage command-line tool.
Because it's based on an event loop, threads are actually very cheap in Lwt when compared to preemptive system threads. Sleeping actually registers an event that will wake up the associated thread when possible. Depending on the backend, the event registering slightly differs.

## Ex.1: Sleep and Join

Write a program that spins off two threads, each of which sleep for some amount of time, say 1 and 2 seconds and then one prints "Heads", the other "Tails". After both have finished, it prints "Finished" and exits. For the time being, use `Lwt_unix.sleep` and `Lwt_io.write` to sleep and print to the console respectively. When we convert this to a Mirage unikernel we will replace these calls (and indeed, any calls that depend on the host OS) with calls to their Mirage replacements, which are guaranteed to exist on all supported platforms (notably Unix and Xen).

### Solution

First, we'll open the `Lwt` module and we'll define a couple of helper functions:

````ocaml
open Lwt ;;

(* boilerplate for `iocaml` -- will be replaced by Mirage *)
let sleep = Lwt_unix.sleep ;;
let log c = prerr_endline ;;
````

Next, we'll define the actual Lwt thread that we wish to run, using Lwt functions directly:

````ocaml
let start c =
    (* the actual guts of the program *)
    bind (join [
        bind (sleep 1.0) (fun () ->
            log c "Heads"; return ()
            );
        bind (sleep 2.0) (fun () ->
            log c "Tails"; return ()
            );
        ]) 
        (fun () ->
            log c "Finished"; return ()
        )
````

Finally, we'll kick off the main Lwt thread via `Lwt_main.run`:

````ocaml
Lwt_main.run (start ())
````

As expected, we see the output with `Tails` tpyically displayed __after__ `Heads` due to the (probably) longer `sleep`.

## Syntax Extensions

Using Lwt does sometimes require significantly restructuring code, and in particular doesn't work with many of the more imperative OCaml control structures such as `for` and `while`. Luckily, Lwt includes a comprehensive [pa_lwt](http://ocsigen.org/lwt/api/Pa_lwt) syntax extension that makes writing threaded code as convenient as vanilla OCaml. Mirage includes this extension by default, so you can use it anywhere you want.

This is a good place to introduce some of these extensions. When opening the Lwt module, the infix operator `>>=` is made available. This operator is an alternative to the `bind` function and often makes the code more readable. E.g. consider `bind (bind (bind t f) g) h` and the operator based equivalent expression `t >>= f >>= g >>= h`. We can now rewrite the previous solution more simply:

````ocaml
let start c =
    join [
        (sleep 1.0 >>= fun () -> (log c "Heads"; return ()));
        (sleep 2.0 >>= fun () -> (log c "Tails"; return ()));
       ] 
    >>= fun () ->
         log c "Finished"; return ()    
````

````ocaml
Lwt_main.run (start ())
````

Further readability enhancements can be made using the `Lwt` syntax extension, e.g., using the anonymous bind, `x >> y`, in place of `x >>= fun () -> y` or the `Lwt`-binding

```ocaml
lwt x = e1 in e2
```

which is equivalent to the `>>=` operator:

```ocaml
e1 >>= fun x -> e2
```

However, idiomatic usage in Mirage now tends to avoid use of syntax extensions generally. Even without use of syntax extensions, further improvements can be made though:

````ocaml
let start c =
    let heads = 
        sleep 1.0 >>= fun () -> 
        return (log c "Heads")
    in
    let tails =
        sleep 2.0 >>= fun () -> 
        return (log c "Tails")
    in
    heads <&> tails
    >>= fun () ->
        return (log c "Finished")
````

````ocaml
Lwt_main.run (start ())
````

## Cancelling Threads

It is often necessary to be able to put time limits on blocking operations. Doing so requires the ability to `cancel` a thread once a timeout has expired. This can be done using `Lwt.cancel 'a Lwt.t`:

````ocaml
open Lwt

let timeout f t =
    sleep f >>= fun () -> cancel t; return ()
````

## Ex.2: Timeouts

However, this example doesn't allow the value return by thread `t` to be used if available. Modify the `timeout` function to rectify this, so it returns

+ `None` if `t` has not yet returned after `f` seconds, or
+ `Some v` if `t` returns `v` within `f` seconds. In order to achieve this behaviour it is possible to use the function state that, given a thread, returns the state it is in, either Sleep, Return or Fail.

### Solution

This solution and application are found in [`lwt/src/unikernels.ml`](https://github.com/mirage/mirage-skeleton/blob/master/lwt/src/unikernels.ml) in the repository. Build with target `timeout1-build`.

````ocaml
let timeout f t =
    sleep f >>= fun () ->
    match state t with
    | Return v -> return (Some v)
    | _        -> cancel t; return None
````

Does your solution match the one given here and always returns after `f` seconds, even when `t` returns within `f` seconds?

This is a good place to introduce a third operation to compose threads: `pick`.

```ocaml
  val pick : 'a t list -> 'a t
```

`pick` behaves exactly like `choose` except that it cancels all other sleeping threads when one terminates.

## Ex.3: Cancelling the Timeout

In a typical use of a timeout, if `t` returns before the timeout has expired, one would want the timeout to be cancelled right away. The next challenge is to modify the timeout function to return `Some v` right after `t` returns. Of course if the timeout does expire then it should cancel `t` and return `None`.

In order to test your solution, you can compile it to a mirage executable and run it using the skeleton provided for the first challenge.

### Solution

Can be found in `lwt/src/unikernels.ml` in the repository. The target is `timeout2-build`.

````ocaml
let timeout f t =
    let tmout = sleep f in
    pick [
        (tmout >>= fun () -> return None);
        (t >>= fun v -> return (Some v));
    ]
````

Write an echo server, reading from a dummy input generator and, for each line it reads, writing it to the console. The server should never stop listening to the dummy input generator. Here is a basic dummy input generator:

## Using Mailboxes

Among the different modules the Lwt library provides is `Lwt_mvar`. This module eases inter-thread communication. Any thread can place a value in a mailbox using the `put` function; dually, the `take` function removes a value from a mailbox and returns it. `take`'s type, `'a Lwt_mvar.t -> 'a Lwt.t`, indicates that a call to the function may block (and let other threads run). The function actually returns only when a value is available in the mailbox.

Here are the needed functions from the `Lwt_mvar` module:

```ocaml
  (* type of a mailbox variable *)
  type 'a t

  (* creates a new empty mailbox variable *)
  val create_empty : unit -> 'a t

  (* puts a value into a mailbox variable, and blocks if it is full *)
  val put : 'a t -> 'a -> unit Lwt.t

  (* will take any available value and block if the mailbox is empty *)
  val take : 'a t -> 'a Lwt.t
```

### Ex.5: Pipeline Parallelism

Write a small set of functions to help do pipeline parallelism. The interface to be implemented is the following (names should give away the appropriate semantic):

```ocaml
  val map: ('a -> 'b Lwt.t) -> 'a Lwt_mvar.t -> 'b Lwt_mvar.t
  val split : ('a * 'b) Lwt_mvar.t -> 'a Lwt_mvar.t * 'b Lwt_mvar.t
  val filter: ('a -> bool Lwt.t) -> 'a Lwt_mvar.t -> 'a Lwt_mvar.t
```

### Solution

````ocaml
let map f m_in =
  let m_out = Lwt_mvar.create_empty () in
  let rec aux () =
    Lwt_mvar.(
      take m_in   >>=
      f           >>= fun v ->
      put m_out v >>= fun () ->
      aux ()
    )
  in
  let t = aux () in
  m_out

let split m_ab =
  let m_a, m_b = Lwt_mvar.(create_empty (), create_empty ()) in
  let rec aux () =
    Lwt_mvar.take m_ab >>= fun (a, b) ->
    join [
      Lwt_mvar.put m_a a;
      Lwt_mvar.put m_b b;
    ] >>= fun () -> aux ()
  in
  let t = aux () in
  (m_a, m_b)

let filter f m_a =
  let m_out = Lwt_mvar.create_empty () in
  let rec aux () =
    Lwt_mvar.take m_a >>= fun a ->
    f a >>= function
      | true -> Lwt_mvar.put m_out a >>= fun () -> aux ()
      | false -> aux ()
  in
  let t = aux () in
  m_out
````

This code can also be found in [`lwt/src/mvar_unikernels.ml`](https://github.com/mirage/mirage-skeleton/blob/master/lwt/src/unikernels.ml) in the [`mirage-skeleton`](https://github.com/mirage/mirage-skeleton/) repository.

Note that in each of the above a recursive Lwt thread is created and will run forever. However, if the pipeline ever needs to be torn down then this recursive thread should be cancelled. This can be done by modifying the above functions to also return the `'t Lwt.t` returned by `aux` in each case, which can then be cancelled when required.

More details can be found in the full [`Lwt` tutorial](http://openmirage.org/wiki/tutorial-lwt).
